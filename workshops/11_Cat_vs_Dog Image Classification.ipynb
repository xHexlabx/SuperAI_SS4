```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os
```

```python
#get-data
train_path = '/content/drive/MyDrive/Datasets/catvsdog/src/train/train'
test_path = '/content/drive/MyDrive/Datasets/catvsdog/src/test/test'
```

```python
#model
inputs = tf.keras.layers.Input( shape = (224 , 224 , 3))

x = tf.keras.layers.Conv2D(6 , (3 , 3) , padding = 'same' , activation = 'relu')(inputs)
x = tf.keras.layers.MaxPool2D()(x)
x = tf.keras.layers.Conv2D(16 , (3 , 3) , padding = 'same' , activation = 'relu')(x)
x = tf.keras.layers.MaxPool2D()(x)
x = tf.keras.layers.Conv2D(36 , (3 , 3) , padding = 'same' , activation = 'relu')(x)
x = tf.keras.layers.MaxPool2D()(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(128 , activation = 'relu')(x)
x = tf.keras.layers.Dense(128 , activation = 'relu')(x)
x = tf.keras.layers.Dense(2 , activation = 'softmax')(x)

model = tf.keras.Model(inputs , x)
model.compile( loss='sparse_categorical_crossentropy' , optimizer = 'adam' , metrics = ['accuracy'])
```

```python
model.summary()
```

```python
images = []

for file in os.listdir(train_path) :

    images.append(file)

images = np.array(images)
np.random.shuffle(images)
```

```python
epochs = 5
batch_size = 512

x_train = []
y_train = []

for epoch in range(epochs) :

    print(epoch)

    for idx , file in enumerate(images) :

        if idx % batch_size == 0 and idx != 0 :

            x_train = np.array(x_train) / 255.0
            y_train = np.array(y_train)

            model.fit(x_train , y_train  )

            x_train = []
            y_train = []

        image = np.array(plt.imread(os.path.join(train_path , file)))
        image = tf.image.resize(image , (224 , 224))

        if file.startswith('cat') :
            x_train.append(image)
            y_train.append([0])

        else :
            x_train.append(image)
            y_train.append([1])
```
